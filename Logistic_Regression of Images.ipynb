```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metric
```

```python
%matplotlib Inline
digits=load_digits()
```

```python
print(digits.data.shape)
digits.target.shape
```

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:7],digits.target[0:7])):
    plt.subplot(1,8,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title("Target Variable Images%i\n"%label,fontsize=10)
    
    
    
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.4,random_state=2)
```

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title("Training plot %i\n"%label,fontsize=20)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(solver='lbfgs', max_iter=10000)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
score=lr.score(X_test,y_test)
print(score)
```

```python
from sklearn import metrics
```

```python
cm=metrics.confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt='.3f',lw=2,square=True,cmap='BuGn_r')
plt.ylabel('Actual label')
plt.ylabel('Predicted Label')
all_sample_title='Actual Accuracty of model {}'.format(score)
plt.title(all_sample_title,size=15)
```
